### Root Mean Squared Error 
$$
RMSE = \sqrt{ \frac{ \sum\nolimits_{i=1}^{n} ( Y_i - \widehat{Y_i})^2 }{n} }
$$

RMSE measures the square root of the average of squared differences between actual and predicted values. It gives an idea of how much error is in the predictions.

### Mean Absolute Error 
$$
MAE = \frac{\sum\nolimits_{i=1}^{n} |Y_i - \widehat{Y_i}|}{n}
$$

MAE measures the average of the absolute differences between actual and predicted values. It represents the average magnitude of errors in a set of predictions, without considering their direction.

### Mean Absolute Percentage Error 
$$
MAPE = \frac{1}{n} \sum_{i=1}^{n} \left| \frac{Y_i - \widehat{Y_i}}{Y_i} \right|
$$

MAPE calculates the percentage difference between the actual and predicted values. It provides an error measure in percentage terms, making it easier to interpret in real-world contexts.

### R-Squared 
$$
R^2 = 1 - \frac{\sum\nolimits_{i=1}^{n} (Y_i - \widehat{Y_i})^2}{\sum\nolimits_{i=1}^{n} (Y_i - \overline{Y})^2}
$$

R-squared measures the proportion of the variance in the dependent variable that is predictable from the independent variables. A value closer to 1 indicates that the model explains a larger proportion of the variance in the data.

Where:
- $n$ = Total Number of Data<br>
- $i$ = Variable i<br>
- $Y_i$ = Actual Data<br>
- $\widehat{Y_i}$ = Prediction Data
- $\overline{Y_i}$ = Mean of actual data values

```python
results = pd.DataFrame(columns=["Feature", "Statistic", "p-value", "Normality"])

def analyze_distribution(data, feature):
    stat, p_value = shapiro(data[feature].dropna())
    p_value_formatted = f"{p_value:.4f}"
    normality_result = "normal" if p_value >= 0.05 else "not normal"
    results.loc[len(results)] = [feature, stat, p_value_formatted, normality_result]

for feature in features:
    analyze_distribution(df, feature)

results
```

## Choosing the Right Scaler and Encoder

### Scaler Selection

| Scenario | Best Scaler |
|----------|------------|
| Data is **normally distributed** | `StandardScaler` |
| Data has **a fixed range & no outliers** | `MinMaxScaler` |
| Data has **outliers** | `RobustScaler` |
| Data is **sparse** | `MaxAbsScaler` |
| Using **Neural Networks** | `MinMaxScaler` |
| Using **PCA** or **SVM** | `StandardScaler` |

### Encoding Selection

| Scenario | Best Encoder |
|----------|-------------|
| Small categories (<10) | `One-Hot Encoding` |
| Many categories (>50) | `Binary Encoding` or `Target Encoding` |
| Tree-based models (Random Forest, XGBoost) | `One-Hot Encoding` or `Target Encoding` |
| Linear models (Logistic Regression, Linear Regression) | `Binary Encoding` or `Target Encoding` |
| Preventing Overfitting | `One-Hot Encoding (if small)` or `Binary Encoding` |
| High-cardinality categorical data | `Target Encoding` |


```python
# Mapping digunakan untuk ordinal encoder
coverage_mapping = [{'col' : 'Coverage',
                   'mapping' : {None : 0, 'Basic' : 1,'Extended' : 2, 'Premium' : 3}}]
rot_mapping = [{'col' : 'Renew Offer Type',
                'mapping' : {None : 0, 'Offer1' : 1,'Offer2' : 2, 'Offer3' : 3, 'Offer4' : 4}}]
education_mapping = [{"col" : "Education",
                      'mapping' : {None : 0, 'High School or Below' : 1,'College' : 2, 'Bachelor' : 3, 'Master' : 4, 'Doctor' : 5}}]


preprocessor = ColumnTransformer(
    transformers=[
    ('onehot', OneHotEncoder(drop = 'first'), ['Vehicle Class', 'EmploymentStatus', 'Marital Status']), # Onehot encoder untuk data nominal
    ('coverage_ordinal', ce.OrdinalEncoder(mapping = coverage_mapping), ['Coverage']),                   # Ordinal encoder untuk data ordinal
    ('rot_ordinal', ce.OrdinalEncoder(mapping = rot_mapping), ['Renew Offer Type']),                     # ---------------//-----------------
    ('education_ordinal', ce.OrdinalEncoder(mapping = education_mapping), ['Education']),
    ('scaler', RobustScaler(), ['Number of Policies', 'Monthly Premium Auto', 'Total Claim Amount']),
    ('scaler2', MaxAbsScaler(), ['Income'])
], remainder = "passthrough")
```

>For Model Scoring we will use neg_mean_absolute_error instead of mean_absolute_error because:

>**mean_absolute_error:** If you use the default mean_absolute_error with a function like cross_val_score, scikit-learn would return higher MAE values as "better," which is opposite to what we want (we want smaller MAE values to indicate better performance).

>**neg_mean_absolute_error:** This inverts the MAE scale so that lower error values result in higher scores when the function maximizes the score. The closer to 0 the negative MAE gets, the better the

```python
from sklearn.ensemble import RandomForestRegressor

# Definisi model
models = {
    'RandomForest': RandomForestRegressor(random_state=42),
    'KNN': KNeighborsRegressor(),
    'DecisionTree': DecisionTreeRegressor(),
    'XGBoost': XGBRegressor(objective='reg:squarederror', random_state=42),
    'LinearRegression': LinearRegression()
}

# Hyperparameter grid untuk setiap model
param_grid = {
    'RandomForest': {
        'model__n_estimators': [100, 200, 300],
        'model__max_depth': [3, 5, 10, None],
        'model__min_samples_split': [2, 5, 10],
        'model__min_samples_leaf': [1, 2, 4]
    },
    'KNN': {
        'model__n_neighbors': [3, 5, 7, 9],
        'model__weights': ['uniform', 'distance'],
        'model__metric': ['euclidean', 'manhattan']
    },
    'DecisionTree': {
        'model__max_depth': [3, 5, 10, None],
        'model__min_samples_split': [2, 5, 10],
        'model__min_samples_leaf': [1, 2, 4]
    },
    'XGBoost': {
        'model__n_estimators': [100, 200, 300],
        'model__max_depth': [3, 5, 10],
        'model__learning_rate': [0.01, 0.1, 0.2]
    },
    'LinearRegression': {
    }
}

# Dictionary untuk menyimpan GridSearchCV
best_models = {}

# Mulai pencarian grid
start_time = time.time()

for name, model in models.items():
    pipeline = Pipeline([
        ('preprocessor', preprocessor),  # Pastikan 'preprocessor' sudah didefinisikan
        ('model', model)
    ])
    
    gscv = GridSearchCV(
        pipeline, 
        param_grid[name], 
        cv=KFold(n_splits=5, shuffle=True, random_state=42), 
        scoring=["neg_root_mean_squared_error", "neg_mean_absolute_error", "neg_mean_absolute_percentage_error"], 
        n_jobs=-1, 
        refit="neg_mean_absolute_error"
    )
    
    print(f"Training {name}...")
    gscv.fit(X_train, y_train)  # FIT ke data pelatihan
    
    best_models[name] = gscv  # Simpan model ke dictionary

end_time = time.time()

# Waktu fitting
fitting_time = end_time - start_time
print(f"Total fitting time: {fitting_time:.2f} seconds")

# Validasi apakah semua model berhasil disimpan di best_models
print("\nModels trained successfully:", best_models.keys())

# Menampilkan score cross-validasi untuk semua model
cross_val_results = []

for name, gscv in best_models.items():
    if not hasattr(gscv, "cv_results_"):  # Cek apakah cv_results_ ada
        print(f"Warning: Model {name} does not have cv_results_. Skipping...")
        continue
    
    results = pd.DataFrame(gscv.cv_results_)
    
    cross_val_results.append({
        "Model": name,
        "Mean RMSE": -results["mean_test_neg_root_mean_squared_error"].mean(),
        "STD RMSE": results["std_test_neg_root_mean_squared_error"].mean(),
        "Mean MAE": -results["mean_test_neg_mean_absolute_error"].mean(),
        "STD MAE": results["std_test_neg_mean_absolute_error"].mean(),
        "Mean MAPE": -results["mean_test_neg_mean_absolute_percentage_error"].mean(),
        "STD MAPE": results["std_test_neg_mean_absolute_percentage_error"].mean()
    })
```

>- Suggestion: It is recommended that binning techniques be implemented for features with a high number of outliers, particularly in the context of variables such as Monthly Premium Auto and Number of Policies.
```python
def vc_input():
    vc_dict = {"1" : "Four-Door Car",
               "2" : "Two-Door Car",
               "3" : "SUV",
               "4" : "Sports Car",
               "5" : "Luxury SUV",
               "6" : "Luxury Car"}
    
    print("Vehicle Class Options:")
    for i in vc_dict.keys():
        print(f"{i}. {vc_dict[i]}")
    vc = input("Choose Vehicle Class (1-6): ")
    if vc in vc_dict:
        return vc_dict[vc]
    else:
        return vc_input()

def cov_input():
    cov_dict = {"1" : "Basic",
                "2" : "Extended",
                "3" : "Premium"}
    
    print("Coverage Options:")
    for i in cov_dict.keys():
        print(f"{i}. {cov_dict[i]}")
    cov = input("Choose Coverage (1-3): ")
    if cov in cov_dict:
        return cov_dict[cov]
    else:
        return cov_input()

def rot_input():
    rot_dict = {"1" : "Offer1",
                "2" : "Offer2",
                "3" : "Offer3",
                "4" : "Offer4"}
    
    print("Renew Offer Type Options:")
    for i in rot_dict.keys():
        print(f"{i}. {rot_dict[i]}")
    rot = input("Choose Renew Offer Type (1-4): ")
    if rot in rot_dict:
        return rot_dict[rot]
    else:
        return rot_input()

def emp_input():
    emp_dict = {"1" : "Employed",
                "2" : "Unemployed",
                "3" : "Medical Leave",
                "4" : "Disabled",
                "5" : "Retired"}
    
    print("Employment Status Options:")
    for i in emp_dict.keys():
        print(f"{i}. {emp_dict[i]}")
    emp = input("Choose Employment Status (1-5): ")
    if emp in emp_dict:
        return emp_dict[emp]
    else:
        return emp_input()

def mar_input():
    mar_dict = {"1" : "Single",
                "2" : "Married",
                "3" : "Divorced"}
    
    print("Marital Status Options:")
    for i in mar_dict.keys():
        print(f"{i}. {mar_dict[i]}")
    mar = input("Choose Marital Status (1-3): ")
    if mar in mar_dict:
        return mar_dict[mar]
    else:
        return mar_input()

def edu_input():
    edu_dict = {"1" : "High School or Below",
                "2" : "College",
                "3" : "Bachelor",
                "4" : "Master",
                "5" : "Doctor"}
    
    print("Education Options:")
    for i in edu_dict.keys():
        print(f"{i}. {edu_dict[i]}")
    edu = input("Choose Education (1-5): ")
    if edu in edu_dict:
        return edu_dict[edu]
    else:
        return edu_input()

def nop_input():
    nop = float(input("Choose Number of Policies (1-9): "))
    if 1 <= nop <= 9:
        return nop
    else:
        return nop_input()

def mpa_input():
    mpa = float(input("Choose Monthly Premium Auto (61-297): "))
    if 61 <= mpa <= 297:
        return mpa
    else:
        return mpa_input()

def tca_input():
    tca = float(input("Choose Total Claim Amount (0.423310-2759.794354): "))
    if 0.423310 <= tca <= 2759.794354:
        return tca
    else:
        return tca_input()

def inc_input():
    inc = float(input("Choose Income (0-99934): "))
    if 0 <= inc <= 99934:
        return inc
    else:
        return inc_input()

def clv_input():
    clv = float(input("Choose Customer Lifetime Value (1898.007675-83325.381190): "))
    if 1898.007675 <= clv <= 83325.381190:
        return clv
    else:
        return clv_input()
```
